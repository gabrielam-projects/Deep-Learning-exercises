```python
import torch
```

```python
# Definir variables x e y como tensores que requieren gradiente
x = torch.tensor(2.0, requires_grad=True)
y = torch.tensor(3.0, requires_grad=True)

# Definir la función z = x^2 + y^2
z = x**2 + y**2

# Realizar la retropropagación para calcular los gradientes
z.backward()

# Ver los gradientes calculados
print(f'Gradiente de x: {x.grad}')  # Derivada de z respecto a x
print(f'Gradiente de y: {y.grad}')  # Derivada de z respecto a y
```

```python
import torch
from torchviz import make_dot

# Crear tensores con requires_grad=True para rastrear el grafo computacional
x = torch.tensor(2.0, requires_grad=True)
y = torch.tensor(3.0, requires_grad=True)

# Definir una operación: z = x^2 + y^2
z = x**2 + y**2

# Usar torchviz para generar el grafo computacional
grafo = make_dot(z, params={'x': x, 'y': y})

# Visualizar el grafo (esto guarda la imagen en un archivo PDF o la muestra en línea)
grafo.render("grafo_computacional", format="png")
grafo
```
